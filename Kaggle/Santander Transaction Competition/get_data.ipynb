# 50% of the data is fake...we're gonna extract the real samples from the test set...

## If a value is unique in its whole column, then that particular example is real...

```python
col_names = [f'var_{i}' for i in range(200)]
for col in tqdm(col_names):
    count = test[col].value_counts()
    uniques = count.index[count==1] # Get all where count is one...
    test[col + "_u"] = test[col].isin(uniques) # True False column
    
    
test["has_unique"] = test[[col + "_u" for col in col_names]].any(axis=1) # Atleast one of them should have a unique
                                                                         # value in any of the features...
```

```python
test["has_unique"]
```

```python
test["has_unique"].sum()
```

```python
real_test = test.loc[test["has_unique"], ["ID_code"] + col_names]
fake_test = test.loc[~test["has_unique"], ["ID_code"] + col_names]
train_and_test = pd.concat([train, real_test], axis=0)
```
